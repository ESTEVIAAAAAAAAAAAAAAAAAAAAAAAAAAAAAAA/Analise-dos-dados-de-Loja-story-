```python
!pip install openpyxl
import pandas as pd 

df = pd.read_excel(r"C:\Users\stays\OneDrive\Desktop\dadosdaloja.xlsx")
df.head()
```

```python
df['VALOR_TOTAL'] = df['QUANTIDADE'] * df['PREÇO_UNITÁRIO']
```

```python
vendas_por_cidade =  df.groupby('CIDADE') ['VALOR_TOTAL'].sum().sort_values(ascending=False)
vendas_por_cidade.plot(kind='bar' , title='Total de Vendas  por Cidade' , ylabel ='R$')
```

```python
mais_vendidos = df.groupby('PRODUTO') ['QUANTIDADE'].sum().sort_values(ascending=False)
mais_vendidos.plot(kind='barh' , title='Produtos mais vendidos por (Quantidade)')
```

```python
top_receita = df.groupby('PRODUTO') ['VALOR_TOTAL'].sum().sort_values(ascending=False).head(5)
top_receita.plot(kind="pie" , autopct='%1.1f%%', title='Top 5 produtos por  Receita')
```
